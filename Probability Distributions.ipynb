```python
import scipy.stats
import numpy as np
import seaborn as sns, matplotlib.pyplot as plt
sns.set_theme()

import pymc as pm
```

```python
pdf = np.linspace(0,1,100)
fig, ax = plt.subplots()
for x in [1,2,5,10]:
    y = scipy.stats.beta.pdf(x=pdf, a=1, b=x)
    plt.plot(pdf, y, label=f"Beta(1,{x})")
fig.legend()
fig.suptitle("Beta Distributions")
```
