```python
import gibbs.distributions as dist
from gibbs.sampler import sampler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
sns.set_theme()
```

```python
mu = dist.Normal(5, 10)
tau = dist.Gamma(1, 1)
```

```python
y = np.random.normal(5, 3, size=100)
sns.displot(x=y, kind='hist', kde=True)
```

```python
mu_, tau_ = sampler(mu, tau, y, samples=10000, thinning=100)
```

```python
fig, ax = plt.subplots()
sns.lineplot(y=mu_.estimates, x=range(0,len(mu_.estimates)))
ax.hlines(y=[y.mean()], xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1])
ax.hlines(y=[mu_.estimates.mean()], xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1], color="red")
```

```python
fig, ax = plt.subplots()
sns.lineplot(y=tau_.estimates, x=range(0,len(tau_.estimates)))
ax.hlines(y=[1/y.std()], xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1])
ax.hlines(y=[tau_.estimates.mean()], xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1], color="red")
```
